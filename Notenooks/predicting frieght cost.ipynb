```python
X=vendor_df[['Quantity','Dollars']]
y=vendor_df['Freight']
```

```python
vendor_df.describe().round()
```

```python
from sklearn.model_selection import train_test_split

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from  sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
model1=LinearRegression()
model1.fit(X_train,y_train)

model2=DecisionTreeRegressor(random_state=42)
model2.fit(X_train,y_train)

model3=RandomForestRegressor(random_state=42)
model3.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(model, X_test, y_test, model_name):
    
    preds = model.predict(X_test)

    mae = mean_absolute_error(y_test, preds)

    # RMSE calculation
    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)

    r2 = r2_score(y_test, preds) * 100

    print(f"\n{model_name} Performance:")
    print(f"MAE  : {mae:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"R2   : {r2:.2f}%")
```

```python
evaluate_model(model1,X_test,y_test,'Linear Regression')
evaluate_model(model2,X_test,y_test,'Decision Tree Regressor')
evaluate_model(model3,X_test,y_test,'Random Forest Regressor')
```

```python
import matplotlib.pyplot as plt

# Get predictions for the Random Forest model (model3)
preds = model3.predict(X_test)

# Plot Actual vs Predicted values
plt.figure(figsize=(6, 6))
plt.scatter(y_test, preds, color='#7f1e5a', alpha=0.6, label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual Freight Cost')
plt.ylabel('Predicted Freight Cost')
plt.title('Actual vs Predicted Freight Cost (Random Forest)')
plt.legend()
plt.show()
```
